```python
import pandas as pd
data = {'Pat1':[1,2,3,4,5], 'Pat2':[5,4,3,2,1], 'Pat3':[3,2,5,4,1], 'Pat4':[1,3,2,5,4]}
df = pd.DataFrame(data)

def Stand():
    for Pat in df.columns:
        Pat_col_sum = df[Pat].sum()
        df[Pat] = df[Pat]/Pat_col_sum

Stand()
df
```

```python
import pandas as pd
import numpy as np
data = {'Pat1':[1,2,3,4,5], 'Pat2':[5,4,3,2,1], 'Pat3':[3,2,5,4,1], 'Pat4':[1,3,2,5,4]}
df = pd.DataFrame(data)

def Shan():
    for Pat in df.columns:
        Pat_col_sum = df[Pat].sum()
        df[Pat] = df[Pat]/Pat_col_sum
        df[Pat] = -1*df[Pat]*(np.log(df[Pat]))
    df.loc['Total']= df.sum()
Shan()
df
```
